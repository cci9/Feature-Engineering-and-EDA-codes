```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_train = pd.read_excel(r"flight_traindata.xlsx")
df_test = pd.read_excel(r"flight_testdata.xlsx")
```

```python
df = pd.concat([df_train, df_test])
```

```python
df.head()
```

```python
df["DurationHour"]=df["Duration"].str.split(" ").str[0].str.split("h").str[0]
```

```python
df.head()
```

```python
df["DurationHour"].unique()
```

```python
df[df["DurationHour"]=="5m"]
```

```python
df = df[(df["Duration"]!="5m")] #other way could be df.drop(rownumber, axis=0, inplace=True)
```

```python
df["DurationHour"].unique()
```

```python
df["DurationHour"] = df["DurationHour"].astype(int)
```

```python
df["DurationMin"]=df["Duration"].str.split(" ").str[1].str.split("m").str[0]
```

```python
df.head()
```

```python
df["DurationMin"]=df["DurationMin"].astype(int)
```

```python
df["DurationMin"]=df["DurationMin"].fillna(0)
```

```python
df["DurationMin"]=df["DurationMin"].astype(int)
```

```python
df.head()
```

```python
df.drop("Duration", axis=1, inplace=True)
```

```python
df["Duration"]=df["DurationHour"]*60+df["DurationMin"]
```

```python
df.head()
```

```python
df.drop(["DurationHour", "DurationMin"], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.drop("Route", axis=1, inplace=True)
```

```python
df.info()
```

```python
df["DOT_Date"]=df["Date_of_Journey"].str.split("/").str[0]
df["DOT_Month"]=df["Date_of_Journey"].str.split("/").str[1]
df["DOT_Year"]=df["Date_of_Journey"].str.split("/").str[2]

df["DOT_Date"]=df["DOT_Date"].astype(int)
df["DOT_Month"]=df["DOT_Month"].astype(int)
df["DOT_Year"]=df["DOT_Year"].astype(int)
```

```python
df.head()
```

```python
df.drop("Date_of_Journey", axis=1, inplace=True)
```

```python
df["DepHr"]=df["Dep_Time"].str.split(":").str[0]
df["DepMin"]=df["Dep_Time"].str.split(":").str[1]

df["DepHr"]=df["DepHr"].astype(int)
df["DepMin"]=df["DepMin"].astype(int)
df.drop("Dep_Time", axis = 1, inplace = True)
```

```python
df["ArrHr"]=df["Arrival_Time"].str.split(" ").str[0].str.split(":").str[0]
df["ArrMin"]=df["Arrival_Time"].str.split(" ").str[0].str.split(":").str[1]
```

```python
df.head()
```

```python
df["ArrHr"]=df["ArrHr"].astype(int)
df["ArrMin"]=df["ArrMin"].astype(int)
df.drop("Arrival_Time", axis = 1, inplace = True)
```

```python
df.head()
```

```python
df["Total_Stops"].unique()
```

```python
df["Total_Stops"]=df["Total_Stops"].map({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3, '4 stops':4})
```
